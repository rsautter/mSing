```python
import NMFDFA
import matplotlib.pyplot as plt
import numpy as np
import chi
```

```python
rn = chi.cNoise(2,(256,256))
plt.figure()
plt.imshow(rn, cmap='gray')
plt.show()
```

```python
wn = chi.cNoise(0,(256,256))
plt.figure()
plt.imshow(wn, cmap='gray')
plt.show()
```

```python
q=np.linspace(-2,10,100)
lag, f = NMFDFA.NMDFA(rn, np.array([8,16,32,64]), q)
```

```python
a, fa = NMFDFA.singularity_spectrum(lag,f,q)
```

```python
plt.figure()
plt.plot(a, fa, 'o-')
plt.xlabel('a')
plt.ylabel('f(a)')
plt.title('Singularity Spectrum')
plt.grid()
plt.show()
```
